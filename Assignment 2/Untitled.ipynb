```python
'''Kaggle: IBM Attrition
Uncover the factors that lead to employee attrition and explore important questions such as:
1. Show a breakdown of distance from home by job role and attrition.
2. Compare average monthly income by education and attrition'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Employee_Attrition = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

Employee_Attrition
```

```python
#Show a breakdown of distance from home by job role and attrition

Dist_by_JR_Att = Employee_Attrition[["DistanceFromHome", "JobRole","Attrition"]]
Dist_by_JR_Att.head()
```

```python
#Plot

#Dist_by_JR_Att.groupby(['JobRole', 'Attrition']).plot.bar()

Dist_by_JR_Att.groupby(["JobRole", "Attrition"])["DistanceFromHome"].sum()
```

```python
#2. Compare average monthly income by education and attrition

Dist_by_JR_Att.groupby(["JobRole", "Attrition"])["DistanceFromHome"].mean()
```

```python
#2. Compare average monthly income by education and attrition

MInc_by_Ed_Att = Employee_Attrition[["MonthlyIncome", "Education","Attrition"]]
MInc_by_Ed_Att.head()
```

```python
MInc_by_Ed_Att.groupby(["Education", "Attrition"])["MonthlyIncome"].mean()
```
